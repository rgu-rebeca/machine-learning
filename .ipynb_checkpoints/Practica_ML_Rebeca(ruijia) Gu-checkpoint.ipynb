## 3. Hacer limpieza: eliminar columnas inutiles

Primero, elimino las columnas que son IDs, y las columnas que son descripciones con textos largos que no sirven para la prediccion de precio.

```python

columns_id = [c for c in train_data.columns if "id" in c.lower()]
train_data = train_data.drop(columns=columns_id)
```

```python
columns_id = [c for c in test_data.columns if "id" in c.lower()]
test_data = test_data.drop(columns=columns_id)
```

```python
train_data.select_dtypes(exclude='object').isna().mean().sort_values(ascending=False) * 100
```

Suprimo la columna de Square Feet, la de Weekly Price, Monthly Price, Security Deposit y Cleaning Fee ya que tiene un porcentaje muy alto de valores nulos y que ademas son variables que no son de tipo object. En cuanto a las variavles de tipo object, realizare posteriormente un proceso de limpieza para eliminarlas, ya que la mayoria contiene textos largos que no aportan valor al analisis. No obstante, antes de eliminarlas, aplicare encoding a las dos variables categoricas mas relevantes: city y room type. 


```python
train_data = train_data.drop(columns=[
    'Square Feet',
    'Weekly Price',
    'Monthly Price',
    'Security Deposit',
    'Cleaning Fee'
])
```

```python
test_data = test_data.drop(columns=[
    'Square Feet',
    'Weekly Price',
    'Monthly Price',
    'Security Deposit',
    'Cleaning Fee'
])
```

```python
train_data.select_dtypes(exclude='object').isna().mean().sort_values(ascending=False) * 100
```

## Imputo los valores nulos de las columnas

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

num_cols = train_data.select_dtypes(include=['float64', 'int64']).columns

train_data[num_cols] = imputer.fit_transform(train_data[num_cols])
test_data[num_cols]  = imputer.transform(test_data[num_cols])
```

Aqui solo hago transform con el test para evitar data leakage.

```python
train_data.select_dtypes(exclude='object').isna().mean().sort_values(ascending=False) * 100
```

## Preparar los datos para el modelo

```python
y_train = train_data['Price']
X_train = train_data.drop(columns=['Price'])

feature_names = X_train.columns
y_test = test_data['Price']
X_test = test_data.drop(columns=['Price'])
```

## Hacer Random Forest para obtener los 10 variables mas importantes. Random Forest sirve para feature pre-screening porque maneja bien relacion no lineal, no sufre por colinearidad, y estima importancia de variables. 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

maxDepth = range(1,15)
param_grid = {'max_depth': maxDepth }
grid = GridSearchCV(DecisionTreeRegressor(random_state=0), param_grid=param_grid, cv = 3, verbose=2)
grid.fit(X_train, y_train)
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

scores = np.array(grid.cv_results_['mean_test_score'])
plt.plot(maxDepth,scores,'-o')
plt.xlabel('max_depth',fontsize=16)
plt.ylabel('3-fold R cuadrado')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

maxDepthOptimo = grid.best_params_['max_depth']
randomForest = RandomForestRegressor(max_depth=maxDepthOptimo,n_estimators=200,max_features='sqrt').fit(X_train,y_train)

importances = randomForest.feature_importances_
importances = importances / np.max(importances)

indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,10))
plt.barh(range(X_train.shape[1]),importances[indices])
plt.yticks(range(X_train.shape[1]),feature_names[indices])
plt.show()
```

## Como indica el resultado de random forest, cogemos los primeros 10 variavles mas importante para entrenar el modelo. 

```python
important_features = ['City_encoded', 'Bedrooms', 'Accommodates', 'Bathrooms', 'Room Type_Private room', 'Beds',
                      'Latitude', 'Longitude', 
                      'Guests Included', 'Host Listings Count']
X_train_selected = X_train[important_features]
X_test_selected = X_test[important_features]
```

## Primero pruebo con un modelo facil, que es el Regresion Lineal

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_selected)
X_test_scaled = scaler.transform(X_test_selected)
# normalizo con los datos de train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

alpha_vector = np.logspace(-1,10,20)
param_grid = {'alpha': alpha_vector }
grid = GridSearchCV(Lasso(), scoring= 'neg_mean_squared_error', param_grid=param_grid, cv = 5, verbose=2)
grid.fit(X_train_scaled, y_train)
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

#-1 porque es negado
scores = -1*np.array(grid.cv_results_['mean_test_score'])
plt.semilogx(alpha_vector,scores,'-o')
plt.xlabel('alpha',fontsize=16)
plt.ylabel('5-Fold MSE')
plt.show()
```

```python
best_lasso = grid.best_estimator_
```

```python
y_train_pred = best_lasso.predict(X_train_scaled )
```

```python
y_test_pred = best_lasso.predict(X_test_scaled )
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
train_r2 = r2_score(y_train, y_train_pred)
print("Train RMSE:", rmse)
print("Train R²:", train_r2)
```

```python

rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2 = r2_score(y_test, y_test_pred)
print("RMSE:", rmse)
print(f"R²: {r2:.3f}")
w = best_lasso.coef_
for f,wi in zip(important_features,w):
    print(f,wi)
```

El modelo de linear regression da un R2 de 0.421 del test, significa que este modelo explica solo el 42.1% de la variabilidad de la variable objetivo,que es el precio. El Test RMSE es 54.65, o sea este modelo se equivoca unos 54.65 € de media

## Probar con un modelo mas complicado: Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

maxDepth = range(1,15)
tuned_parameters = {'max_depth': maxDepth}

grid = GridSearchCV(RandomForestRegressor(random_state=0, n_estimators=200, max_features='sqrt'), param_grid=tuned_parameters,cv=3, verbose=2) 
grid.fit(X_train_scaled, y_train)

print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

scores = np.array(grid.cv_results_['mean_test_score'])
plt.plot(maxDepth,scores,'-o')
plt.xlabel('max_depth')
plt.ylabel('CV score')

plt.show()
```

```python
best_rf = grid.best_estimator_
```

```python
tree = best_rf.estimators_[0] 
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plot_tree(
    tree,
    feature_names=X_train.columns,
    filled=True,
    rounded=True,
    max_depth=3  # limitar profundidad por la legibilidad
)
plt.show()
```

```python
from sklearn.tree import export_text

print(
    export_text(
        tree,
        feature_names=important_features,
        max_depth=5
    )
)
# esta manera de visualizarlo en texto me parece mas legible
```

```python
y_train_pred_rf = best_rf.predict(X_train_scaled)
```

```python
y_test_pred_rf = best_rf.predict(X_test_scaled)
```

```python
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred_rf))
train_r2 = r2_score(y_train, y_train_pred_rf)
print(f"Train RMSE: {train_rmse:.2f}")
print("Train R²:", train_r2)
```

```python


rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_rf))
r2 = r2_score(y_test, y_test_pred_rf)

print(f"Test RMSE: {rmse:.2f}")
print(f"Test R²: {r2:.3f}")
```

El modelo de random forest explica un 59.6% de la variabilidad del variable objetivo (el precio). No esta mal el R cuadrado. El Test RMSE es 45.68, o sea este modelo se equivoca unos 45.68 € de media

## Conclusion: El modelo de Random Forest predice mejor el precio de las viviendas de Airbnb en comparacion con el modelo de regresion lineal. Esto es razonable, ya que la relacion entre el precio y las variables explicativas no suele ser lineal. Ademas, Random Forest es capaz de capturar relaciones no lineales e interacciones entre variables, lo que le permite adaptarse mejor a la complejidad del problema.
